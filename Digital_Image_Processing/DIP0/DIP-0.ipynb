```python
# the difference between pillow and cv

# The first subplot displays the image processed in Pillow library and is labeled as 'Pillow'.
plt.subplot(1,2,1)
plt.title('Pillow')
plt.imshow(color_pencil_pillow)
plt.axis('off')

# The second subplot displays the image processed in the OpenCV (CV) library and is labeled as 'CV'.
plt.subplot(1,2,2)
plt.title('CV')
plt.imshow(color_pencil_cv)
plt.axis('off')
```

#### let's check what exactly happened!
In OpenCV, images are represented as matrices with different color channels. The three main color channels are red, green, and blue (RGB). 

The statement "in the opencv red channel replaced by blue channel and blue channel replaced by red channel" means that the image was modified in a way that the red channel data were replaced with the blue channel data, and the blue channel data were replaced with the red channel data. In other words, the colors in the red channel were swapped with the colors in the blue channel, and vice versa. 

This modification can have various effects on the image depending on its original content and purpose. For instance, if the image had a lot of red objects or details, they may appear blue after this modification, and vice versa.

```python
# Load the image 'colors.png' using OpenCV and Pillow
cv = cv2.imread(r'images/colors.png')
pilow = Image.open(r'images/colors.png')

# Create subplots to display the images side by side
# The first plot shows the Pillow image
plt.subplot(1, 2, 1)
plt.imshow(pilow)
plt.title('Pilow') # Set the title of the plot
plt.axis('off') # Hide the axis from the plot

# The second plot shows the OpenCV image
plt.subplot(1, 2, 2)
plt.imshow(cv)
plt.title('CV') # Set the title of the plot
plt.axis('off') # Hide the axis from the plot
```

```python
# fixing the problem
# The following line of code fixes the problem by converting the color space of 'cv' from BGR to RGB
fixed_cv = cv2.cvtColor(cv, cv2.COLOR_BGR2RGB)

# Displaying the original and fixed images side by side using subplots
plt.subplot(1,2,1)
plt.title('Previous CV')
plt.imshow(cv)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Fixed CV')
plt.imshow(fixed_cv)
plt.axis('off')
```

## Grayscale and color images

Color images have three channels (R for red, G for green, and B for blue) while grayscale images have only one channel. To convert a color image to grayscale, you can use a formula that weights the contribution of each channel, such as taking the average of the RGB values or using a specific set of weights, such as 0.299 for red, 0.587 for green, and 0.114 for blue. This formula is applied to every pixel in the image to produce a single gray value.     
gray=(red+green+blue)/3    
Or     
gray=0.299red+0.587green+0.114blue.

```python
# Reading image in color mode
color_image = cv2.imread(r'images/color_pencil.jpg') # reads the image file 'golden.jpg' in color mode

# Converting image to grayscale
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY) # converts the image from color mode to grayscale mode

# Saving grayscale image
cv2.imwrite('gray.png', gray_image) # saves the grayscale image with the filename 'gray.png'

# Displaying grayscale image using matplotlib
plt.title('CV') # sets the title of the displayed image as 'CV'
plt.imshow(gray_image, cmap='gray') # displays the grayscale image using a grayscale colormap
plt.axis('off') # turns off the display of axis values on the plot
```

```python
# read color image as an array(BGR)
img_channels = cv2.imread(r'images/colors.png')

# accessing the blue channel in the color image
# read all rows and columns for the first (0th) channel
blue_ch = img_channels[:,:,0]

# accessing the green channel in the color image
# read all rows and columns for the second (1st) channel
green_ch = img_channels[:,:,1]

# accessing the red channel in the color image
# read all rows and columns for the third (2nd) channel
red_ch = img_channels[:,:,2]

# displaying channels and the original image using plotting

# converting the OpenCV format image to Matplotlib format(RBG)
con_img_channels = cv2.cvtColor(img_channels, cv2.COLOR_BGR2RGB)


# the first plot shows the original image
plt.subplot(2, 2, 1)
plt.imshow(con_img_channels)
plt.title('original img')
plt.axis('off')

# the second plot shows the blue channel
plt.subplot(2, 2, 2)
plt.imshow(blue_ch, cmap='gray')
plt.title('blue ch')
plt.axis('off')

# the third plot shows the green channel
plt.subplot(2, 2, 3)
plt.imshow(green_ch, cmap='gray')
plt.title('green ch')
plt.axis('off')

# the fourth plot shows the red channel
plt.subplot(2, 2, 4)
plt.imshow(red_ch, cmap='gray')
plt.title('red ch')
plt.axis('off')
```
